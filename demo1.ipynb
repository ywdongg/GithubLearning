```python
# prompt: 我想要使用matplotilib畫出sin()產生的正弦波和cos()產生的餘弦波，疊在一張表格上

import matplotlib.pyplot as plt
import numpy as np

# 建立 x 座標的數值範圍
x = np.linspace(0, 2 * np.pi, 100)

# 計算 sin(x) 和 cos(x)
y_sin = np.sin(x)
y_cos = np.cos(x)

# 建立圖表
plt.plot(x, y_sin, label='sin(x)')
plt.plot(x, y_cos, label='cos(x)')

# 設定圖表標題和軸標籤
plt.title('正弦波和餘弦波')
plt.xlabel('x')
plt.ylabel('y')

# 顯示圖例
plt.legend()

# 顯示圖表
plt.show()
```
